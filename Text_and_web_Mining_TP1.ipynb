 TF :  gives us the frequency of the word in each document in the corpus. It is the ratio of number of times the word appears in a document compared to the total number of words in that document. It increases as the number of occurrences of that word within the document increases. Each document has its own tf.

$TF = (Frequency  of  the word in the sentence) / (Total number of words in the sentence)$

```python
doc1 = "You are trying to code TF-IDF all by yourself like a big girl / boy"
doc2 = "So this is a tinny doc"
doc3= "And another tinny doc to test few stuff."
doc4 = "So in total, we are four documents, have fun ;)."
```

```python
bagOfwords1 = doc1.split(' ')
bagOfwords2 = doc2.split(' ')
bagOfwords3 = doc3.split(' ')
bagOfwords4 = doc4.split(' ')
```

```python
uniqueWords = set(bagOfwords1).union(set(bagOfwords2), set(bagOfwords3), set(bagOfwords4))
uniqueWords
```

```python
#creat the dictionaries to keep words count
numOfWords1 = dict.fromkeys(uniqueWords, 0)
numOfWords2 = dict.fromkeys(uniqueWords,0)
numOfWords3 = dict.fromkeys(uniqueWords, 0)
numOfWords4 = dict.fromkeys(uniqueWords, 0)
```

```python
#count the words in bag of words (bagOfwords1,bagOfwords2,bagOfwords3,bagOfwords4)
for word in bagOfwords1:
    numOfWords1[word] += 1
    
for word in bagOfwords2:
    numOfWords2[word] += 1
    
for word in bagOfwords3:
    numOfWords3[word] += 1
    
for word in bagOfwords4:
    numOfWords4[word] += 1
     
```

```python
df = pd.DataFrame([numOfWords1,numOfWords2,numOfWords3,numOfWords4])
df
```

```python
def ComputeTF(wordDict, bow):
    # tf = (nbr of accuenrce of the word in the D/ nbr of words in D)
    tfDict = {}
    nbrOfWords = len(bow)
    for word, count in wordDict.items():
        tfDict[word] = count/float(nbrOfWords)
    return tfDict
```

```python
Tf1 = ComputeTF(numOfWords1, bagOfwords1)
Tf2 = ComputeTF(numOfWords2, bagOfwords2)
Tf3 = ComputeTF(numOfWords3, bagOfwords3)
Tf4 = ComputeTF(numOfWords4, bagOfwords4)
Tf1["boy"]
```

# Inverse Data Frequency (IDF)


$IDF(t) = log_e(Total number of documents / Number of documents with term t in it).$ 

```python
def ComputeIDF(documents):
    N = len(documents)
    #compter le nombre de document qui contient le mot 
    idfDict = dict.fromkeys(documents[0].keys(), 0)
    for document in documents:
        for word, val in document.items():
            if val > 0:
                idfDict[word] += 1
    
    for word, val in idfDict.items():
        idfDict[word] = math.log(N / float(val))
    return idfDict
    
    
```

```python
idfs = ComputeIDF([numOfWords1, numOfWords2, numOfWords3, numOfWords4])
idfs["boy"]
```

# TF-IDF

the TF-IDF is simply the TF multiplied by IDF :
       <img src="TF-IDF.png">

```python
def ComputeTFIDF(tfBagOfWords, idfs):
    tfidf = {}
    for word, val in tfBagOfWords.items():
        tfidf[word] = val * idfs[word]
    return tfidf
```

```python
tfidf1 = ComputeTFIDF(Tf1, idfs)
tfidf2 = ComputeTFIDF(Tf2, idfs)
tfidf3 = ComputeTFIDF(Tf3, idfs)
tfidf4 = ComputeTFIDF(Tf4, idfs)
tfidf1["boy"]
```
